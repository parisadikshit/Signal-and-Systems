# Fourier Transform

```python
import numpy as np
from scipy.fftpack import fft,fftfreq,rfft
from scipy  import integrate
import matplotlib.pyplot as plt
```

```python
# fft of another signal
t = np.arange(0,1,0.01)
```

```python
a = np.sin(2*np.pi*10*t)+np.sin(2*np.pi*100*t)
plt.grid()
plt.plot(t,a,'r')
plt.title('Input signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
a = np.sin(2*np.pi*10*t)+np.sin(2*np.pi*100*t)
b = fft(a)
c = (abs(b))/len(a)
plt.grid()
plt.plot(c,'r')
plt.title('FFT of Input signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
```

```python
arr = np.array([2,2,2,2,1,1,1])
arr_y = np.fft.fft(arr)
arr1 = abs(arr_y)
print(arr_y)
plt.grid()
plt.plot(np.arange(7),arr1,'bo')
plt.title('FFT of signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
```

```python
abs(np.fft.ifft(arr_y)) # the array that was previously given obtained again
```

```python
x = np.linspace(-10,10,num=50)
x.shape[-1]
```

```python
# sine signal
y = np.sin(x)
plt.grid()
plt.plot(x,y,'bo')
plt.title('Input signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
# fft of sine 
x = np.linspace(-20,20,num=50)
y = np.fft.fft(np.sin(x))
freq = np.fft.fftfreq(x.shape[-1])
plt.grid()
plt.plot(freq,y.real,freq,y.imag)
plt.title('FFT of Input signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
```

```python
# fft of sine
x = np.linspace(-10,10,num=50)
y = fft((np.sin(x))).real
plt.grid()
plt.plot(x,y,'bo',x,y,'b')
plt.title('FFT of Input signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
```

```python
#exponential of -x
x = np.linspace(-10,10,num=50)
y = np.exp(-x)
plt.grid()
plt.plot(x,y,'ro',x,y,'b')
```

```python
#fft of exponential of -x
x = np.linspace(-10,10,num=50)
y = fft(np.exp(-x)).real
plt.grid()
plt.plot(x,y,'ro',x,y,'b')
plt.title('FFT of exp(-x)')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
```

```python
# fft of gaussian function 
t = np.arange(-1,1,10)
s = np.exp(-np.pi*t*t)
```

```python
t = np.arange(10)
s = fft(np.exp(-np.pi*t*t))
plt.grid()
plt.plot(t,s.real,'ro',t,s.real,'r')
plt.title('FFT of Input signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
```
