# Fourier Series Representation

For given signal,it is given:$ F_0=4 Hz,  \tau= \frac{T}{2} seconds, A=1.0V $ and Number of coefficients=30


 $c_n = \frac{A\tau}{T} * \frac{sin(n*pi*F_0*\tau)}{n*pi*F_0*\tau}$ for n!=0 and $c_n =\frac{A\tau}{T}  $

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Fo = 4
T = 1/Fo
tau = T/2
A = 1
N = 32
wo = 2*np.pi*Fo
c=[]
for k in range(1,N//2):
    c.append((A*tau/T)*((np.sin(k*np.pi*Fo*tau)/(k*np.pi*Fo*tau))))
```

```python
plt.stem(c,use_line_collection=True)
```

```python
# value of tau is halved here
Fo = 4
T = 1/Fo
tau = T/4
A = 1
N = 32
wo = 2*np.pi*Fo
k=0
c=[]
for k in range(1,N//2):
    c.append((A*tau/T)*((np.sin(k*np.pi*Fo*tau)/(k*np.pi*Fo*tau))))
```

```python
plt.stem(c,use_line_collection=True)
plt.xlabel('Frequency')
plt.ylabel('Cn')
plt.show()
```

```python
C=[abs(i) for i in c]
plt.stem(C,use_line_collection=True)
plt.xlabel('Frequency')
plt.ylabel('|Cn|')
plt.show()
```

```python

plt.stem(np.angle(c),use_line_collection=True)
plt.xlabel('Frequency')
plt.ylabel('0n')
plt.show()
```
